# Bare Iron Slab

```python
ind = 4

plot_name = "01_fe_graph/dos/pl_pdos_bands_Fe_slab.html"
plot_title = "FCC Iron (111)"
pdos_filter_dict = {
#     "type": ["sum-up", "sum-down"],
#     "atom_ind": [3],
#     "atom_ind": [11]  
#     "band": ["d"],
#     "element": ["Fe"],
#     "type": ["sum"],
    }

entry = df.loc[ind]
pdos_data = entry.pdos_data
bands_data = entry.bands_data
# view(entry.init_atoms)

# Band Analysis
bands_data, bands_layout = plot_bands(bands_data)

# PDOS & DOS Analysis
dos_data, pdos_data_out, dos_layout = plot_pdos_dos(
    pdos_data,
    entry.init_atoms,
    filter_dict=pdos_filter_dict,
    group="band",
    e_range=[-1, 1],
    plot_title="N-doped Graphene Supported on FCC Fe(111)",
    )

# del dos_layout["xaxis"]["range"]

fig = plot_pdos_bands(
    pdos_data_out,
    bands_data,
    dos_layout,
    bands_layout,
    plot_title=plot_title,
    )
py.plotly.iplot(fig, filename=plot_name)
```
