### PDOS Analysis

## N-doped Graphene Supported on FCC Fe(111)

```python
entry = df.loc[11]
pdos_data = entry.pdos_data
atoms = entry.init_atoms

# view(atoms)

filter_dict = {
#     "type": ["sum-up", "sum-down"],
#     "atom_ind": [6, 7, 8, 9, 10, 11, 12, 13, 14],
#     "atom_ind": [11]  
    "band": ["d"],
    "element": ["Fe"],
#     "type": ["sum"],
    }

dos_data, pdos_data_out, layout = plot_pdos_dos(
    pdos_data,
    filter_dict,
    atoms,
    plot_title="N-doped Graphene Supported on FCC Fe(111)",
    )

del layout["xaxis"]["range"]

py.plotly.iplot(
    {
        "data": pdos_data_out,
        "layout": layout,
        },
    filename="__temp__/pl_pdos_N_graph_Fe.html"
    )
```
